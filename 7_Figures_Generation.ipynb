# Graphs (With Explanation) **7 Figures Generation.**




Using:
* Matplotlib
* Seaborn

```python
import matplotlib.pyplot as plt   # Used for plotting graphs
import seaborn as sns             # Used for advanced visualization
import numpy as np                # Used for generating data
```


<font color="red" size="5"><b>1. Line Plot</b></font>


```python
x = [1,2,3,4]     # Data for x-axis
y = [10,20,25,30] # Data for y-axis

plt.plot(x, y)                 # Create line plot
plt.title("Line Plot")         # Set title
plt.xlabel("X-axis")           # Label x-axis
plt.ylabel("Y-axis")           # Label y-axis
plt.show()                     # Display the graph
```

<font color="red" size="5"><b>2. Bar Chart</b></font>

```python
categories = ['A','B','C']   # Categories
values = [5,7,3]             # Values

plt.bar(categories, values)  # Create bar chart
plt.title("Bar Chart")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

<font color="red" size="5"><b>3. Histogram</b></font>

```python
data = np.random.randn(100)   # Generate 100 random values

plt.hist(data)                # Create histogram
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

<font color="red" size="5"><b>4. Scatter Plot</b></font>

```python
x = [1,2,3]
y = [4,5,6]

plt.scatter(x, y)            # Create scatter plot
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

<font color="red" size="5"><b>5. Box Plot</b></font>

```python
data = [10,20,30,40,50]      # Dataset

sns.boxplot(x=data)          # Create box plot
plt.title("Box Plot")
plt.show()
```

<font color="red" size="5"><b>6. Heatmap</b></font>

```python
data = np.random.rand(5,5)   # Create 5x5 matrix

sns.heatmap(data, annot=True)  # Create heatmap with values
plt.title("Heatmap")
plt.show()
```

<font color="red" size="5"><b>7. Pie Chart</b></font>

```python
values = [10,20,30]           # Data values
labels = ['A','B','C']        # Labels

plt.pie(values, labels=labels, autopct='%1.1f%%')  # Create pie chart
plt.title("Pie Chart")
plt.show()
```
