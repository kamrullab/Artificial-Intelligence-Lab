## To perform numerical operations using NumPy.

```python
import numpy as np

# Create array
arr = np.array([1, 2, 3, 4])

# Operations
print("Array:", arr)
print("Mean:", np.mean(arr))
print("Sum:", np.sum(arr))

# Matrix
matrix = np.array([[1,2],[3,4]])
print("Matrix:\n", matrix)
```
