### To learn basic Python concepts such as variables, loops, and functions.

```python
# Declare two variables
a = 10
b = 5

# Perform addition and display result
print("Sum:", a + b)   # print() is used to show output

# Loop to print numbers from 0 to 4
for i in range(5):     # range(5) generates 0,1,2,3,4
    print("Number:", i)

# Define a function
def square(x):         # x is the input parameter
    return x * x       # returns square of x

# Call the function
print("Square:", square(4))
```
